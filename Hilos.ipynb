```python
from re import sub
import threading
import random
import time

def ordena_subvector(subvector,hilo):
  tiempo_inicio = time.time()
  subvector.sort()
  tiempo_fin = time.time()
  tiempo_ejecucion = tiempo_fin - tiempo_inicio
  print(f"Hilo {hilo}: Subvector ordenado (Tiempo:{tiempo_ejecucion}) ")



def dividir_vector(vector,num_hijos):
  longitud_subvector = len(vector)
  subvectores = [vector[i:i+ longitud_subvector] for i in range (0,len(vector), longitud_subvector)]
  return subvectores

def unir_vectores(subvectores):
  vector_ordenado = [num for subvector in subvectores for num in subvector]
  return vector_ordenado



def ordena_vector(vector,num_hijos):
  subvectores = dividir_vector(vector,num_hijos)
  threads = []

  for i, subvector in enumerate(subvectores):
    thread = threading.Thread(target= ordena_subvector, args=(subvector,i))
    thread.start()
    threads.append(thread)
    for thread in threads:
      thread.join()
    vector_ordenado = unir_vectores(subvectores)
    print(f"vector ordenado final: {vector_ordenado}")


vector_grande = [random.randint(1,10000) for _ in range(10000)]

num_hilos = int(input("Ingrese la cantidad de hilos"))


ordena_vector(vector_grande,num_hilos)

#No hay problema en la distribucion de los hilos ya que los segmentos del vector son asignados a cada hilo
#Este problema esta diseñado para ser resuelto con este enfoque




```
